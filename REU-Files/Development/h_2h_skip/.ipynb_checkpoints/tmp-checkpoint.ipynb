```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

files = ['betas.out','pulse.out','wfn.out','fourier.out','hyd2.out','overlap.out','hydpartion.out']

for file_name in files:
    print file_name
    if file_name == 'pulse.out' or file_name=='hydpartion.out':
        test = pd.read_csv("run/"+file_name, delim_whitespace=True)
        correct = pd.read_csv("run/test/"+file_name, delim_whitespace=True)
    else:
        test = pd.read_csv("run/"+file_name, header=None, delim_whitespace=True)
        correct = pd.read_csv("run/test/"+file_name, header=None, delim_whitespace=True)
    plotting = (correct-test).abs()
    print plotting.max()
    plt.figure()
    plotting.plot(ylim=[0,1.1e-6])
    plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
    plt.show()
```
